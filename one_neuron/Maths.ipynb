```python
def plot_vec(vectors, colors):
    plt.figure(dpi =100)
    plt.axhline(y =0, color= 'lightgrey')
    plt.axvline(x =0, color= 'lightgrey')
    plt.xlim(-5,5)
    plt.ylim(-5, 5)
    for i in range(len(vectors)):
        x = np.concatenate([[0,0], vectors[i]])
        plt.quiver(x[0], x[1], x[2], x[3], scale = 1, scale_units= 'xy', angles='xy', color = colors[i])
```

### Image compression using SVD

```python
img = Image.open(os.path.join('..', 'datasets','plants.jpg'))
_ = plt.imshow(img)
```

"1" 1 bit Black and White\
"L" 8 bit Black and White\
"LA" Alpha mode

```python
im_gray = img.convert('LA')
_ = plt.imshow(im_gray)
```

```python
im_mat = np.array(list(im_gray.getdata(band = 0)), float)
im_mat
```

```python
im_mat.shape
```

```python
im_mat.shape = (im_gray.size[1], im_gray.size[0])
im_mat.shape
```

```python
im_mat = np.matrix(im_mat)
_ = plt.imshow(im_mat, cmap = 'gray')
```

```python
# svd method
U, sigma, V = np.linalg.svd(im_mat)
```

```python
# 1st decomposed image
new_img = np.matrix(U[:, :1] * np.diag(sigma[:1]) * np.matrix(V[:1, :]))
_ = plt.imshow(new_img, cmap = 'gray')
```

```python
for i in [2, 4, 8, 16, 32, 64, 128]:
    new_img = np.matrix(U[:, :i] * np.diag(sigma[:i]) * np.matrix(V[:i, :]))
    _ = plt.imshow(new_img, cmap = 'gray')
    plt.title(i)
    plt.show()
```

### Moore-Penrose Pseudoinverse

```python
A = np.array([[2, 3, 6], [5, 6, 4]])
A
```

```python
U, d, V_T = np.linalg.svd(A)
```

```python
U
```

```python
d
```

```python
V_T
```

```python
D = np.diag(d)
D
```

```python
D_inv = np.linalg.inv(D)
D_inv
```

```python
# adjust the rows and columns
D_plus = np.concatenate((D_inv, np.array([[0,0]])), axis = 0)
D_plus
```

```python
# A_plus pseudoinverse of A
np.dot(V_T.T, np.dot(D_plus, U.T))
```

```python
# pseudoinverse function
np.linalg.pinv(A)
```

### Automatic differentiation

```python
# scalar
x = torch.tensor(5.)
x
```

```python
# trace the gradient of x
x.requires_grad_()
```

```python
y = x**2 + 2*x
```

```python
# backward pass till x
y.backward()
```

```python
# gradient of y at x = 5 or dy/dx
x.grad
```

```python
# tensor
x = torch.tensor([2, 3, 5.]).requires_grad_()
x
```

```python
y = x**2 + 2*x
```

```python
# in case of tensor
y.sum().backward()
```

```python
x.grad
```

```python
x = tf.Variable(3.)
x
```

```python
with tf.GradientTape() as t:
    t.watch(x) # trace forward pass
    y = 2*x**3 + x
```

```python
# dy/dx
t.gradient(y, x)
```

```python
x = tf.Variable([2, 5, 3.])
x
```

```python
with tf.GradientTape() as t:
    t.watch(x) # trace forward pass
    y = 2*x**3 + x
```

```python
# dy/dx
t.gradient(y, x)
```
